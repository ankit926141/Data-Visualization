```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('house-price.csv')
```

```python
sns.scatterplot(x='Square footage', y='Price', data=df)
plt.title('House Prices vs. Square Footage')
plt.xlabel('Square footage')
plt.ylabel('Price')
```

```python
plt.show()
```
